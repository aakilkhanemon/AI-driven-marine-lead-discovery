```python
from rdkit import Chem

# Define file paths
input_sdf_path = "Marine_Library_Docking_Ready_Master.sdf"
output_sdf_path = "Extracted_Marine_Compounds.sdf"

# The target compound names to filter for
target_names = ["Gracilin A", "Stypoldione", "Lamellarin H", "Dieckol"]

# Initialize the SD supplier and writer
supplier = Chem.SDMolSupplier(input_sdf_path)
writer = Chem.SDWriter(output_sdf_path)

# Counters for tracking
extracted_count = 0
total_processed = 0

print("Starting extraction process...")
print("-" * 40)

for mol in supplier:
    total_processed += 1
    if mol is None:
        # Skip corrupted or unreadable 3D structures
        continue

    # Get the molecule name (usually stored in the _Name property)
    mol_name = mol.GetProp("_Name").strip() if mol.HasProp("_Name") else ""

    # If name is empty, check common alternative SDF properties
    if not mol_name and mol.HasProp("NAME"):
        mol_name = mol.GetProp("NAME").strip()

    # Check if any of our target names are a substring of the molecule's name
    # This ensures we catch analogs/derivatives (case-insensitive)
    is_match = any(target.lower() in mol_name.lower() for target in target_names)

    if is_match:
        writer.write(mol)
        extracted_count += 1
        print(f"Match found and extracted: {mol_name}")

# Close the writer to flush the data to the file
writer.close()

print("-" * 40)
print(f"Extraction complete!")
print(f"Total molecules scanned: {total_processed}")
print(f"Total structures extracted: {extracted_count}")
print(f"Saved to: {output_sdf_path}")
```

```python
from rdkit import Chem

# Define file paths
input_sdf_path = "Marine_Library_Docking_Ready_Master.sdf"
output_sdf_path = "Extracted_Marine_Compounds.sdf"

# Updated target list to perfectly match your file's naming style
target_names = ["gracilin-a", "stypoldione", "lamellarin-h", "dieckol"]

# Initialize the SD supplier and writer
supplier = Chem.SDMolSupplier(input_sdf_path)
writer = Chem.SDWriter(output_sdf_path)

extracted_count = 0
total_processed = 0

print("Starting extraction with fixed naming keys...")
print("-" * 50)

for mol in supplier:
    total_processed += 1
    if mol is None:
        continue

    # Get molecule name
    mol_name = mol.GetProp("_Name").strip() if mol.HasProp("_Name") else ""
    if not mol_name and mol.HasProp("NAME"):
        mol_name = mol.GetProp("NAME").strip()

    # Case-insensitive substring match against the hyphenated names
    is_match = any(target.lower() in mol_name.lower() for target in target_names)

    if is_match:
        writer.write(mol)
        extracted_count += 1
        print(f"Match found and extracted ({extracted_count}/40): {mol_name}")

writer.close()

print("-" * 50)
print("Extraction complete successfully!")
print(f"Total structures extracted: {extracted_count} (Should be 40)")
print(f"Saved to: {output_sdf_path}")
```
