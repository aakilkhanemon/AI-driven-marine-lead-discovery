```python
import torch
import numpy as np
from transformers import AutoTokenizer, EsmForMaskedLM

def evaluate_vaccine_perplexity(sequence):
    """
    Calculates the pseudo-perplexity of a synthetic vaccine sequence
    using the ESM-2 650M parameter masked protein language model.
    """
    print("[INFO] Initializing ESM-2 (650M parameter model)...")
    model_name = "facebook/esm2_t33_650M_UR50D"

    # Load tokenizer and model onto GPU
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = EsmForMaskedLM.from_pretrained(model_name)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    model.eval()

    # Tokenize input sequence
    inputs = tokenizer(sequence, return_tensors="pt")
    input_ids = inputs["input_ids"].to(device)

    # seq_len excludes the special start (<cls>) and end (<eos>) tokens
    seq_len = input_ids.shape[1] - 2
    print(f"[INFO] Sequence successfully tokenized. Length: {seq_len} AA.")
    print("[INFO] Computing token-by-token pseudo-log-likelihood estimation...")

    total_log_likelihood = 0.0

    # Iterative masking loop for pseudo-perplexity calculation
    with torch.no_grad():
        for i in range(1, seq_len + 1):
            masked_input_ids = input_ids.clone()
            original_token = masked_input_ids[0, i].item()

            # Substitute the current amino acid with the <mask> token
            masked_input_ids[0, i] = tokenizer.mask_token_id

            # Compute model logits
            outputs = model(masked_input_ids)
            logits = outputs.logits

            # Apply softmax to isolate probabilities for the specific masked position
            probabilities = torch.nn.functional.softmax(logits[0, i], dim=-1)
            true_token_prob = probabilities[original_token].item()

            # Accumulate the log likelihood of the true native residue
            total_log_likelihood += np.log(true_token_prob)

    # Compute inverse geometric mean of the pseudo-likelihoods
    pseudo_perplexity = np.exp(-total_log_likelihood / seq_len)
    return pseudo_perplexity

# Your exact Mpox multi-epitope vaccine construct
mpox_vaccine_seq = "EAAAKGIINTLQKYYCRVRGGRCAVLSCLPKEEQIGKCSTRGRKCCRRKKEAAAKSSQDLSFIYAAYYIHPILMALAAYHIDPLGTNVAAYDSDDGFYEWAAYFLADKKMTLAAYNLQDLSLYIAAYNLLCHIYSLAAYLTTRYHSIYAAYYTDLVISKIAAYSSDELDQKKAAYLPTSIPLAYGPGPGHPILMALFGIKLPALGPGPGYYIHPILMALFGIKLGPGPGYIHPILMALFGIKLPGPGPGIHPILMALFGIKLPAGPGPGVKSYNYMLLVNRLTEGPGPGTHPLVLIHSRASPRIKKAGLKSSYTNQTLSLVDEDHIHTSDKSKKGARNQRTKPKRTGNDQSQQPIKKDCMVSIDEVASTHDWSTRKKQNKNMFGGKKDQYISARHIKKSEVDYIGQFDMRFLNSLAIHEKFDAFMNKK"

# Run validation layer
final_score = evaluate_vaccine_perplexity(mpox_vaccine_seq)

print("\n" + "="*50)
print("             MANUSCRIPT METRICS METADATA        ")
print("="*50)
print(f"Target Construct: Next-Generation Mpox Subunit Vaccine")
print(f"Sequence Length : {len(mpox_vaccine_seq)} Amino Acids")
print(f"ESM-2 Pseudo-Perplexity Score: {final_score:.4f}")
print("="*50)
```

```python
import json
import numpy as np
import os

# 1. Parse the AlphaFold 3 Server json text parameters
print("[INFO] Parsing AlphaFold 3 json matrix configuration...")
with open("/content/candidate_3_pae.json", "r") as f:
    af3_data = json.load(f)

# Extract the raw 2D coordinate error array matrix
# AlphaFold 3 JSON typically nests this under the 'pae' or 'predicted_aligned_error' key
if "pae" in af3_data:
    pae_matrix = np.array(af3_data["pae"])
elif "predicted_aligned_error" in af3_data:
    pae_matrix = np.array(af3_data["predicted_aligned_error"])
else:
    # Fallback extraction logic for alternative AlphaFold JSON dumps
    key_name = [k for k in af3_data.keys() if "pae" in k.lower() or "error" in k.lower()][0]
    pae_matrix = np.array(af3_data[key_name])

# 2. Convert and save as the exact binary name expected by t2pmhc
target_npy_path = "/content/candidate_3_predicted_aligned_error.npy"
np.save(target_npy_path, pae_matrix)
print(f"[SUCCESS] Binary structural matrix written to: {target_npy_path}")

# 3. Re-write the samplesheet.tsv to use the newly generated .npy file configuration
import pandas as pd
data = {
    "sample_id": ["candidate_3"],
    "pdb_path": ["/content/candidate_3.pdb"],
    "pae_path": [target_npy_path], # Re-mapped to the standard numpy path variables
    "tcr_beta_cdr3": ["CASSLAPGTTNEKLFF"]
}
df = pd.DataFrame(data)
df.to_csv("samplesheet.tsv", sep="\t", index=False)
print("[SUCCESS] Samplesheet.tsv re-compiled successfully. You can now safely re-run your GNN prediction cell.")
```
